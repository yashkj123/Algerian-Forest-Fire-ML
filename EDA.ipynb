# Algerian Forest Fire Prediction Using Regression

## 1. Introduction
This project predicts the Fire Weather Index (FWI) using weather conditions in Algerian forests. Models used: Linear, Ridge, Lasso, ElasticNet.

## 2. Dataset Overview
- Source: [Algerian Forest Fires dataset]
- Target Variable: `FWI`
- Features include: Temperature, RH, WS, Rain, etc.

## 3. Data Cleaning & Preparation
- Removed non-numeric rows
- Standardized column names
- Saved cleaned CSV

## 4. Feature Selection & Scaling
- Selected all numerical features except FWI
- Applied StandardScaler

## 5. Model Training
Trained Linear, Ridge, Lasso, and ElasticNet regressors using train/test split.

## 6. Evaluation Metrics
Used R² Score and Mean Squared Error to evaluate models.

## 7. Final Model Selection
Ridge was chosen based on highest R² and lowest MSE.

## 8. Model Saving
Saved the Ridge model and Scaler using `pickle`.

## 9. Prediction Function
Function created to make predictions using saved model.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset, skip the first junk row
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", skiprows=1)

# Preview first 5 rows
df.head()
```

```python
#missing values 
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##remove the 122nd row
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

#### Changes the required columns as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

#### Changing the other columns to float data datatype


```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```
